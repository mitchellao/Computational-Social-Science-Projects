### Model 1

For my first model I used a ridge regression model. Ridge regression is a linear regression model and therefore assumes that the relationship between predictors and the target is approximately linear. These models also assume that the errors are normally distributed and independent. While ordinary least squares (OLS) can struggle with multicollinearity, ridge regression partially resolves this issue through regularization. The regularization in ridge regression is achieved by adding a penalty term to the loss function, proportional to the squared magnitude of the coefficients. This penalty discourages large coefficients, thereby reducing the risk of overfitting. This makes ridge regression particularly useful when predictors are highly correlated or when there are many predictors relative to the number of observations. One of the main distinctions between ridge regression and Lasso regression is that ridge regression does not shrink coefficients to zero. Thus, ridge models retain all features. This lack of automatic feature selection may hurt interpretability. 

Overall, ridge regression is preferable to a basic OLS model for this inspection prediction problem because it is less likely to overfit the 2011–2013 data. However, more complex machine learning models may perform better. These models are less sensitive to multicollinearity and do not assume a linear relationship between predictors and the target variable, making them better suited for capturing complex, non-linear patterns in the data.

```python
# train/validation split

# Set seed
np.random.seed(10)

# training and test split 
X_train, X_validate, y_train, y_validate = train_test_split(X, 
                                                    y, 
                                                    train_size = .80, 
                                                    test_size= 0.20,
                                                    stratify=y)
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# identify non-dichotomous to transform
columns = list(X_train.select_dtypes(include=['number']).loc[:, X_train.nunique() > 2])

# use loop to transform training data for only columns we want to transform
for column in columns:
    X_train[column] = scaler.fit_transform(X_train[column].values.reshape(-1, 1)).flatten()

# use loop to transform validation data for only columns we want to transform
for column in columns:
    X_validate[column] = scaler.fit_transform(X_validate[column].values.reshape(-1, 1)).flatten()
```

```python

ridge_reg = Ridge()                                             
ridge_model = ridge_reg.fit(X_train, y_train) 

# hyperparameter tuning
param_grid = {'alpha': np.arange(.1, 1, .1),
              'fit_intercept': [True, False],
              'solver': ['auto', 'svd', 'cholesky', 'lsqr']}

ridge_grid_reg = GridSearchCV(ridge_reg,  
                              param_grid, 
                              cv=3)  

ridge_grid_reg.fit(X_train, y_train)


best_index = np.argmax(ridge_grid_reg.cv_results_["mean_test_score"])  
best_ridge_pred = ridge_grid_reg.best_estimator_.predict(X_validate)   

# print the results  
print(ridge_grid_reg.cv_results_["params"][best_index])
print('Best CV R^2:', max(ridge_grid_reg.cv_results_["mean_test_score"]))
print('Validation R^2:', ridge_grid_reg.score(X_validate, y_validate))
print('Validation RMSE', mean_squared_error(best_ridge_pred, y_validate))
```

```python
# picking the best Ridge model
best_ridge_reg = Ridge(alpha=0.9, fit_intercept =False, solver='lsqr') # making ridge regression with optimized hyperparameters

best_ridge_model = best_ridge_reg.fit(X_train, y_train) # training ridge model with optimized hyperparameters
```

###  Model 2

For my second model I used a random forest. Random forest models are an ensemble learning method that builds multiple decision trees and then aggregates their results. The basic logic of ensemble learning is that grouping 'weak learners' together can create a 'strong learner.' This aggregation happens via majority voting. These models introduce randomness to prevent overfitting. It does this in two ways: (1) bootstrap sampling, where each tree is trained on a different random subset of the data. This selection is done with replacement and is called bagging. (2) At each split in the tree, only a random subset of features are considered. Unlike ridge models, random forests do not assume a linear functional form. Random forests still assume that features are independent, which might not be true. Because random forests can handle non-linear data and do not have big overfitting risks, it is a better option for this inspection problem than ridge regression. One major drawback, however, is more difficulty in interpretation. Of course, it is also more computationally expensive. But because the datasets in this analysis are not very big, this should not be an issue. 


```python
# initialize a random forest classifier
rf_classifier = RandomForestClassifier(
                       # specify parameters
                       n_estimators=100,             
                       criterion='gini',              
                       max_depth=None,                
                       min_samples_split=2,          
                       min_samples_leaf=1,            
                       min_weight_fraction_leaf=0.0,  
                       max_features=None,            
                       max_leaf_nodes=None,           
                       min_impurity_decrease=1e-07,  
                       random_state = 10)  

# specify cross-validation -- because I am doing cross validation I do not need to use the train/validation splits from above
scores = cross_val_score(rf_classifier, 
                         X, 
                         y.ravel(),
                         cv=5)

scores.mean()
```

### Model 3

For my third model, I used adaptive boosting. This is another ensemble decision tree. While random forests grow lots of trees in parallel, adaptive boosting models start with a strong classifier and then iterates on it with weak learners. These weak learners are trained on errors. This means boosting models do well for classifying difficult cases that are not clear-cut. Like random forests, adaptive boosting can handle non-linear data well, so it is also a good option for this inspection problem. Because adaptive boosting focuses on difficult classification cases, it handles mis-balanced data well. Becasue this model is sequential, it is more computationally expensive. Again, given the size of the dataset in this analysis, this expense should not be an issue.

Adaptive boosting would be a better option than random forests if inspection data includes specific establishments that are more difficult to predict than others. If classification difficulty is not a significant elements of this data, however, a random forest model might be a better option. 


```python
# initialize an adaptive boosting classifer 
ada_classifier = AdaBoostClassifier(n_estimators=100)
# calculate accuracy using cross validation -- again, I do not need to use the train/validation split 
scores = cross_val_score(ada_classifier,  
                         X,               
                         y.ravel(),       
                         cv = 5)

# calculate mean score across models 
scores.mean()
```

### Validation Metrics

Be sure to explain which of these metrics you would want to prioritize when conducting predictive auditing in this context and why.

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

Accuracy is the ratio of correctly predicted instances to the total instances. It should be used when you have a balanced dataset and when there is no preference for false positives and false negatives. Because most establishments do not fail their inspection, the dataset in this modeling analysis is not as balanced as accuracy demands, and so it is not the preferable validation metric. 

Precision is the proportion of true positive predictions out of all positive predictions. Precision is the metric used when false positves want to be avoided. This is not as much of an issue in this analysis, as false positives are to be expected. Not every establishment that will be inspected will fail. For this reason precision is not the preferable validation metric for this problem. 

Recall is the proportion of true positives identified out of all actual positives. Recall is used when false negatives want to be avoided. That is, when missing positive cases is costly. The recall metric is useful for this analysis, as establishments that are falsely predicted to safe even though they are not (false negative) might pose risks to their patrons. These risks want to be avoided. 

The F1 metric is the harmonic mean of precision and recall. Thus, it balances both metrics. This validation metric might also be useful for this analysis, as the city cannot afford to conduct infinite inspections and so false positives would ideally be avoided along with false negatives. 


First, I will look at the validation metrics for the ridge regression model. To calculate these metrics I need to establish a threshold to classify a prediction as 1 (failure) or 0 (non-failure). For this treshold, I chose a 0.5 cut-off. As detailed above, I will mostly be paying attention to the recall and F1 metrics. These are respectively 0.60 and 0.72. 

```python
# use the model to make predictions

ridge_pred_continuous = best_ridge_model.predict(X_validate)
ridge_pred = (ridge_pred_continuous >= 0.5).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_validate, ridge_pred)
precision = precision_score(y_validate, ridge_pred, average='binary') 
recall = recall_score(y_validate, ridge_pred, average='binary')
f1 = f1_score(y_validate, ridge_pred, average='binary')

# Display results
print("Ridge Accuracy:", accuracy)
print("Ridge Precision:", precision)
print("Ridge Recall:", recall)
print("Ridge F1 Score:", f1)
```

```python
# Specify scoring metrics
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision_macro',
    'recall': 'recall_macro',
    'f1': 'f1_macro'
}
```

Second, I will look at the validation metrics for the random forest model. These metrics were calculated with cross validation. The average recall was 0.90 and the average F1 was 0.88. The random forest model is preforming substantially better than the ridge regression model. This makes sense given the trade-offs of the two models, as described above. 

```python
# Cross-validation for multiple metrics
scores = cross_validate(rf_classifier, X, y, cv=5, scoring=scoring, return_train_score=False)

# Calculate averages for each metric
average_accuracy = scores['test_accuracy'].mean()
average_precision = scores['test_precision'].mean()
average_recall = scores['test_recall'].mean()
average_f1 = scores['test_f1'].mean()

print("RF Average Accuracy:", average_accuracy)
print("RF Average Precision:", average_precision)
print("RF Average Recall:", average_recall)
print("RF Average F1-score:", average_f1)
```

Third, I will look at the validation metrics for the adaptive boosting classifier. These metrics were calculated with cross validation. The average recall was 0.87 and the average F1 was 0.87. The adaptive boosting model is preforming substantially better than the ridge regression model, but not as well as the random forest model. It makes sense that adaptive boosting is doing better than ridge regression for the same reason it made sense that the random forest model was doing better than the ridge. The adapative boosting model is not doing as well as the random forest model, and this suggests that hard-to-classify cases, which adaptive boosting is particularly well suited to deal with, might not be as relevant in this dataset. 

Overall, the random forest model had the best performance on the relevant validation metrics (i.e., recall and F1). 

```python

# Cross-validation for multiple metrics
scores = cross_validate(ada_classifier, X, y, cv=5, scoring=scoring, return_train_score=False)

# Calculate averages for each metric
average_accuracy = scores['test_accuracy'].mean()
average_precision = scores['test_precision'].mean()
average_recall = scores['test_recall'].mean()
average_f1 = scores['test_f1'].mean()

print("AdaBoost Average Accuracy:", average_accuracy)
print("AdaBoost Average Precision:", average_precision)
print("AdaBoost Average Recall:", average_recall)
print("AdaBoost Average F1-score:", average_f1)
```

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?


First of all many datasets that are used in the social sciences and public policy are unbalanced. If you are only calculating accuracy with an unbalanced dataset, a high accuracy might simply reflect the model's bias toward the over-balanced outcome. For example, if you have a dataset where the outcome is sudden death, and not many of the cases actually have the outcome of 'sudden death,' a model that predicts no instances of 'sudden death' will be very accurate. But this is just reflecting the unbalanced dataset, and maybe the model isn't really doing a very good job at predicting the outcome. Something like recall, in contrast, allows you to measure how well a model predicts actual positive cases. 

In a similar vein, how well we want our model to be at avoiding false positives or false negatives will likely depend on the actual content of the question being asked and how it relates to public policy outcomes. When evaluating ML models we should always keep in mind what the priorities of the prediction are and we should select validation criteria that correspond to these priorities. 

2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decisionmaking and real world behavior?

The most basic way policymakers could address the interplay between algorthmic decisionmaking and real world behavior is by making the algorithm opaque so that establishments could not game the system. This is basically the tactic credit rating agencies use. If people do not know how an algorithm is making its predictions, then they cannot game it. However, opaque algorithms are not always the most democratic or socially just way to govern a society, as they leave little room for accountability. 

Another way to resolve this interplay would be to continuously update the algorithm year after year. This would have several benefits. First, establishments would be less able to game the algorithm. Second, the validity of the model would be more likely to be maintained because it is being adjusted for any interfering factors in between years. Of course, there is a resource limit for how often this sort of updating can be done. Ultimately, this is a trade-off that policymakers would have to deliberate.
